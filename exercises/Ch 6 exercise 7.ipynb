```python
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mp
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X, y = make_moons(n_samples=10000, noise=0.4)
```

```python
plt.scatter(X[:,0], X[:,1], s=y*10, marker='^')
plt.scatter(X[:,0], X[:,1], s=abs(y - 1)*10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
tree = DecisionTreeClassifier()
param_grid = [{'max_leaf_nodes':[23, 24, 25, 26, 27], 'max_depth': [7, 8, 9]}]
grid_search = GridSearchCV(tree, param_grid, cv=5, scoring='accuracy', return_train_score=True)
```

```python
grid_search.fit(X_train, y_train)
```

```python
preds = grid_search.predict(X_test)
```

```python
score = accuracy_score(preds, y_test)
score
```
